```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../data/car_insurance_claims.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
corr = df.corr()
sns.heatmap(corr, cmap="crest")
```

```python
# drop high correlating columns because naive bayes assserts indipendent features
df = df.drop(['length','width','gear_box','ncap_rating','max_torque','max_power'], axis=1)
corr = df.corr()
sns.heatmap(corr, cmap="crest")
```

```python
#ONE HOT ENCODING
y = pd.get_dummies(df[['area_cluster']], prefix='cluster')
y1 = pd.get_dummies(df['segment'], prefix='segment')
y2 = pd.get_dummies(df['fuel_type'], prefix='fuel')
y3 = pd.get_dummies(df['engine_type'], prefix='engine')
y4 = pd.get_dummies(df['transmission_type'], prefix='transmission')
y5 = pd.get_dummies(df['steering_type'], prefix='steering')
y6 = pd.get_dummies(df['fuel_type'], prefix='fuel')

df1 = df.drop(['area_cluster', 'segment', 'fuel_type', 'engine_type', 'transmission_type', 'steering_type', 'fuel_type'], axis=1)
df1 = pd.concat([df1,y,y1,y2,y3,y4,y5,y6], axis=1)
df1.head()
```

```python
#LABEL ENCODING

# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
df1['model'] = labelencoder.fit_transform(df1['model'])
df1['is_esc'] = labelencoder.fit_transform(df1['is_esc'])
df1['is_adjustable_steering'] = labelencoder.fit_transform(df1['is_adjustable_steering'])
df1['is_tpms'] = labelencoder.fit_transform(df1['is_tpms'])
df1['is_parking_sensors'] = labelencoder.fit_transform(df1['is_parking_sensors'])
df1['is_parking_camera'] = labelencoder.fit_transform(df1['is_parking_camera'])
df1['rear_brakes_type'] = labelencoder.fit_transform(df1['rear_brakes_type'])
df1['is_front_fog_lights'] = labelencoder.fit_transform(df1['is_front_fog_lights'])
df1['is_rear_window_wiper'] = labelencoder.fit_transform(df1['is_rear_window_wiper'])
df1['is_rear_window_washer'] = labelencoder.fit_transform(df1['is_rear_window_washer'])
df1['is_rear_window_defogger'] = labelencoder.fit_transform(df1['is_rear_window_defogger'])
df1['is_brake_assist'] = labelencoder.fit_transform(df1['is_brake_assist'])
df1['is_power_door_locks'] = labelencoder.fit_transform(df1['is_power_door_locks'])
df1['is_central_locking'] = labelencoder.fit_transform(df1['is_central_locking'])
df1['is_power_steering'] = labelencoder.fit_transform(df1['is_power_steering'])
df1['is_driver_seat_height_adjustable'] = labelencoder.fit_transform(df1['is_driver_seat_height_adjustable'])
df1['is_day_night_rear_view_mirror'] = labelencoder.fit_transform(df1['is_day_night_rear_view_mirror'])
df1['is_ecw'] = labelencoder.fit_transform(df1['is_ecw'])
df1['is_speed_alert'] = labelencoder.fit_transform(df1['is_speed_alert'])
df1.head()
```

```python
df1.to_csv('../data/car-insurance-preprocessed.csv', index=False) 
```
