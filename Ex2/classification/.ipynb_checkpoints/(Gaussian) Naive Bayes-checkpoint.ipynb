```python
def prior_probability(df, label):
    p = df[label][(df[label] == 1)].count()
    n = df[label].count()
    return [p/n, 1-p/n]

def likelyhood_probability(df, feature,x, label):
    #only considering values with given outcome
    df = df[(df['quality'] == label)]
    mean = df[feature].mean()
    sd = df[feature].std()
    #build gaussian distribution PDF
    prob = (1 / (sd * np.sqrt(2 * np.pi))) * (np.exp(-(1/2) * ((x-mean)/sd)**2))
    return prob


def naive_bayes (df, X):
    features = list(X.columns)
    prior = prior_probability(df, 'quality')
    y_pred = []
    for x in X:
        print(x)
        likelyhood = [1,1]
        for l in range(2):
            for f in range(len(features)):
                print(x[f])
                likelyhood[l] *= likelyhood_probability(df, features[f],x[f],l)
        
        post = [0,0]
        for l in range(2):
            post[l] = likelyhood[l] * prior[l]
        
        y_pred.apped(np.argmax(post))
    return np.array(y_pred)
            
        
    
    
    
    
```

```python
X = df.loc[:, df.columns != 'quality']
naive_bayes(df,X)
```
